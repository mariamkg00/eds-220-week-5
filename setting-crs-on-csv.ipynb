```python
# import necessary libraries
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

#update pandas display option
pd.set_option('display.max.columns', None)
```

We want to make a geopandas geodataframe from our csv data. 

```python
# import csv using pandas
power_plants = pd.read_csv('data/power_plants_epsg4269.csv')
power_plants.head(3)
```

From the 'documentation' we know that the CRS for this file is EPSG: 4269 in lat/lon coordinates. We can use this information to make a new `gpd.GeoDataFrame` from the `pd.DataFrame`. To do this, we use the `gpd.points_from_xy` function:

```python
power_plants = gpd.GeoDataFrame(power_plants, # data for geo-dataframe
                               # specify/ make a geometry column
                               geometry = gpd.points_from_xy(power_plants.longitude,
                                                            power_plants.latitude),
                               #specify CRS
                               crs = 'EPSG:4269'
                               )
power_plants.head()
```

Let's check information about our CRS; what does geographic crs mean? 
- we won't be able to get accurate analysis in 2d data; can make it into a CRS and choosing a projection
- start with geographic CRS and then project it via projection 

```python
# print CRS info
print('is geographic?', power_plants.crs.is_geographic)
print('is projected?', power_plants.crs.is_projected)
print('datum: ', power_plants.crs.datum)
print('ellipsoid:', power_plants.crs.ellipsoid)
```

## Data selection

```python
states = states[(states.region !=9) & (~states.statefp.isin([2,15]))]
#everything thing except states that have statefp 2 or 15 
```

```python
power_plants = power_plants[~power_plants.state.isin('Puerto Rico','Hawaii','Alaska')]
```

```python
states = states[(states.region!=9) & (~states.statefp.isin([2,15]))]
power_plants = power_plants[~power_plants.state.isin(['Puerto Rico','Hawaii','Alaska'])]
```
